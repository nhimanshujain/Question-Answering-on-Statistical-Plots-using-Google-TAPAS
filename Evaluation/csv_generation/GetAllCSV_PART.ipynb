```python
plot_types = dict()
for item in annot:
  if item["type"] not in plot_types:
    plot_types[item["type"]] = [item["image_index"]]
  else:
    plot_types[item["type"]].append(item["image_index"])
```

```python
for k in plot_types:
    print(f"{k} = {len(plot_types[k])}")
```

```python
import collections
map_csv_category = collections.defaultdict(list)
names = [os.path.basename(x) for x in glob.glob("/content/Results_2/*.txt")]
for i in names:
  if int(i[:-4]) in plot_types['hbar_categorical']:
    map_csv_category['hbar_categorical'].append(int(i[:-4]))
  elif int(i[:-4]) in plot_types['vbar_categorical']:
    map_csv_category['vbar_categorical'].append(int(i[:-4]))
  elif int(i[:-4]) in plot_types['dot_line']:
    map_csv_category['dot_line'].append(int(i[:-4]))
  elif int(i[:-4]) in plot_types['line']:
    map_csv_category['line'].append(int(i[:-4])) 
  else:
    continue
```
